```python
import pandas as pd
allData = pd.read_csv('world.csv', usecols=['國家', '日期', '總確診數','新增確診數'], index_col='日期')
allData
```

```python
#將索引變成Datetime資料類型
allData.index = pd.to_datetime(allData.index)
allData.info()
```

```python
#使用loc可以直接輸入所要的日期月份來抓資料
allData.loc['2020-02']
```

```python
taiwan = allData.query('國家=="台灣"')
japan = allData.query('國家=="日本"')
```

```python
taiwan1 = taiwan[['總確診數','新增確診數']]
taiwan1.columns = ['台灣總確診數','台灣新增確診數']
taiwan1
```

```python
japan1 = japan[['總確診數','新增確診數']]
japan1.columns = ['日本總確診數','日本新增確診數']
japan1
```
