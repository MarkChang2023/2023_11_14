```python
import numpy as np
import pandas as pd
scores = np.random.randint(50, 101, size=(50,5))
scores_pd = pd.DataFrame(scores,
                         columns=['國文', '英文', '數學', '地理', '歷史'],
                         index=range(1,51)
                         )
scores_pd
```

```python
def callback1(value:pd.Series):
    return pd.Series([value.max(), value.min(), value.median(), value.mean()],
                     index=['最高分', '最低分', '中間值', '平均值'])

scores_pd.apply(callback1)
```
