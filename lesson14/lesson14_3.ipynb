```python
#下載文字檔
import requests
url:str = 'https://www.moeaea.gov.tw/ECW/populace/opendata/wHandOpenData_File.ashx?set_id=248'
response:requests.Response = requests.get(url)
if response.status_code == 200:
    print("下載成功")
    data:str = response.text
elif response.status_code >= 400:
    print("下載失敗")
elif response.status_code > 200:
    print("下載有問題")
```

```python
from io import StringIO
import csv
with StringIO(data) as file:
    dict_reader = csv.DictReader(file)
    csv_data:list[dict] = list(dict_reader)
print(csv_data)
```

```python
#使用DataFrame顯示資料
import pandas as pd
dataFrame = pd.DataFrame(csv_data, columns=['日期(年)','單位','再生能源發電量合計'])
dataFrame
```

```python
#作業5_1讀取個股日成交資訊.csv，不使用存成csv檔的方式，並使用DataFrame顯示下載的資料
#下載文字檔
import requests
url:str = 'https://raw.githubusercontent.com/roberthsu2003/python/master/%E6%AA%94%E6%A1%88%E5%AD%98%E5%8F%96/%E5%80%8B%E8%82%A1%E6%97%A5%E6%88%90%E4%BA%A4%E8%B3%87%E8%A8%8A.csv'
response:requests.Response = requests.get(url)
if response.status_code == 200:
    print("下載成功")
    data:str = response.text
elif response.status_code >= 400:
    print("下載失敗")
elif response.status_code > 200:
    print("下載有問題")
```

```python
#從內建的io模組裡面匯入StringIO這個類別
#使用with as 建立file物件
from io import StringIO
import csv
with StringIO(data) as file:
    dict_reader = csv.DictReader(file)
    csv_data:list[dict] = list(dict_reader)
```

```python
#使用DataFrame顯示資料
import pandas as pd
dataFrame = pd.DataFrame(csv_data)
dataFrame
```
